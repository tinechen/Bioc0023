```python
taxon=pd.read_table("/content/sample_data/5b99703c-6868-4f03-acaf-4759eabfec30/data/taxonomy.tsv", index_col=0)
```

```python
taxon[['kingdom', 'phylum', 'class', 'order', 'family', 'genus', 'species']]=taxon['Taxon'].str.split(';', expand=True)
```

```python
taxon.columns
```

```python
taxon
```

```python
taxa_results=pd.merge(taxon, rel_table, left_index=True, right_index=True)
```

```python
sumtax=taxa_results.groupby('genus').sum()
```

```python
sumtax=sumtax.drop(['Confidence'], axis=1)
```

```python
sumtax.sum()
```

```python
sumtax=sumtax.T
```

```python
sumtax
```

```python
sumtax=pd.merge(metadata,sumtax, left_index=True, right_index=True)
```

```python
sumtax
```

```python
sumtax['sample-id'] = sumtax.index
```

```python
sumtax.columns
```

```python
df2 = sumtax.filter(regex=' g')
lst=df2.columns.values.tolist()
```

```python
sumtax=sumtax.groupby(['group']).mean()
```

```python
sumtax['group']=sumtax.index
```

```python
taxabar= pd.melt(sumtax,
              id_vars=['group'],
              value_vars=sumtax.filter(regex=' g'),
              var_name= 'Taxa', 
              value_name='Abs')
print (taxabar)
```

```python
fig = px.bar(taxabar, x="group", y="Abs", color="Taxa")
fig.show()
```
