```python
import os
import shutil
current_wkd="/content/ftp.ddbj.nig.ac.jp/ddbj_database/dra/fastq/DRA005"
dir_dst= "/content/test/"
for root, dir, files in os.walk(top=current_wkd):
    for file in files:
        if file.endswith(".bz2"):  #match files that match this extension
            shutil.move(os.path.join(root, file), dir_dst)
```

```python
cd /content/test/
```

##Unzip bz2 files

Here we use a simple for statement to unzip the files, each file ending in bz2 is unzipped

```python
%%bash
for f in *.bz2; do bzip2 -d $f; done
```

##gzip files

The same code is applied to our now unzipped fastq files, but to gunzip them

```python
%%bash
for f in *.fastq; do gzip $f; done
```
